```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import WebDriverException as WDE
from selenium.webdriver.common.action_chains import ActionChains
import time
import pandas as pd

#opciones de navegacion
opciones=webdriver.ChromeOptions()
opciones.add_argument('--start-maximized')
opciones.add_argument('--disable-extensions')
```

```python
website='https://www.epdata.es/datos/cambio-climatico-datos-graficos/447'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=opciones)
driver.get(website)
```

```python
driver.execute_script("window.scrollTo(0,33300)")
time.sleep(15)

try:
    rectas=driver.find_elements(by=By.CLASS_NAME,value='gestiongrafica-barra')
except WDE:
    print('Nada')
    
```

```python
años=list()
paises=list()
indices=list()
for recta in rectas:
    actions = ActionChains(driver)
    actions.move_to_element(recta)
    actions.perform()
    time.sleep(3)
    datos=driver.find_element(by=By.XPATH,value='/html/body/div[1]/div/section/div/div[70]/div/article/div[2]/div/div[5]/table').text
    datos=datos.split('\n')
    print(datos)
    año=datos[0].split(' ')[1]
    pais=datos[0].split(' ')[4]
    indice=datos[1].split(' ')[1]
    años.append(año)
    paises.append(pais)
    indices.append(indice)

```

```python
df=pd.DataFrame({'Pais':paises,'Indice de Riesgo climático':indices,'Año':años})

print(df)
```

```python
df.to_csv('RiesgoClimatico.csv',index=False)
driver.quit()
```
