```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
website = 'https://www.datosmundial.com/co2-por-pais.php' # Página web
page = requests.get(website)
page.status_code # Respuesta de la petición HTTP
```

```python
soup = BeautifulSoup(page.content) # page.content contenido en bruto de la respuesta del servidor
print(soup.prettify()) # Estructura anidada 
```

```python
rows = soup.find('table', attrs={"id":"tabsort"}).find('tbody').find_all('tr')
```

```python
country = []
metano = []
person = []
for row in rows:
    country.append(row.find_all('td')[0].get_text())
    metano.append(row.find_all('td')[5].get_text())
    person.append(row.find_all('td')[6].get_text())
```

```python
df = pd.DataFrame({"Pais":country, "Metano total":metano, "Metano por persona": person})
df.head(5)
```

```python
df.to_csv('Metano.csv') # Convert to csv
```
