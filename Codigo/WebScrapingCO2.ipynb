```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import WebDriverException as WDE
import time
import pandas as pd
import numpy as np
```

```python
#opciones de navegacion
opciones=webdriver.ChromeOptions()
opciones.add_argument('--start-maximized')
opciones.add_argument('--disable-extensions')
```

```python
website='https://datosmacro.expansion.com/energia-y-medio-ambiente/emisiones-co2'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),chrome_options=opciones)
driver.get(website)
```

```python
año=2020
for i in range(0,20):
    try:
        WebDriverWait(driver,15)\
            .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                          'button.close')))\
            .click()
    except WDE:
        print('No se encontró el elemento')

    try:
        WebDriverWait(driver,15)\
            .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                              'button.didomi-components-button didomi-button didomi-dismiss-button didomi-components-button--color didomi-button-highlight highlight-button'.replace(' ','.'))))\
            .click()
    except WDE:
        print('No se encontró el elemento')

    WebDriverWait(driver,15)\
        .until(EC.element_to_be_clickable((By.XPATH,
                                              '/html/body/div[3]/div[1]/div/section/div/article/div/div[4]/div/div[2]/table')))

    datos_columnas=driver.find_element(by=By.XPATH,value='/html/body/div[3]/div[1]/div/section/div/article/div/div[4]/div/div[2]/table')
    datos_columnas=datos_columnas.text
    datos_columnas=datos_columnas.split('\n')[1:-1]

    datos=list()
    variacion=list()
    for i in range(0,len(datos_columnas)-1,2):
        datos.append(datos_columnas[i].split(' [+] '))
        variacion.append(datos_columnas[i+1])

    paises=list()
    datos1=list()
    for i in range(0,len(datos)):
        paises.append(datos[i][0])
        datos1.append(datos[i][1])

    datos2=list()
    co2Total=list()
    co2Kg=list()
    co2Pc=list()

    for i in range(0,len(datos1)):
        datos2.append(datos1[i].split(' '))

    for i in range(0,len(datos2)):
        co2Total.append(datos2[i][0])
        co2Kg.append(datos2[i][1])
        co2Pc.append(datos2[i][2])

    df=pd.DataFrame({'Paises':paises,'CO2 Totales Mt':co2Total,'CO2 Kg/1000$':co2Kg,'CO2 t per capita':co2Pc,'Var':variacion})

    df=df.assign(Año=año)

    print(df)
    if i==0:
        df.to_csv(r'emisionCO2.csv',index=False)
    else:
        df.to_csv(r'emisionCO2.csv',mode='a',index=False)

    año-=1

    WebDriverWait(driver,10)\
        .until(EC.element_to_be_clickable((By.XPATH,
                                              '/html/body/div[3]/div[1]/div/section/div/article/div/div[4]/div/div[1]/span[1]/a')))\
        .click()

driver.quit()
```
